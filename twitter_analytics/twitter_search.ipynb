```python
import os
import sys
import json
from requests_oauthlib import OAuth1Session
from dotenv import load_dotenv
```

```python
# .env
load_dotenv()

# 
if os.getenv("CONSUMER_KEY") is None or os.getenv("CONSUMER_SECRET") is None or os.getenv("ACCESS_TOKEN") is None or os.getenv("ACCESS_TOKEN_SECRET") is None:
    sys.exit(".env")

# OAuth
twitter = OAuth1Session(os.getenv("CONSUMER_KEY"), client_secret=os.getenv("CONSUMER_SECRET"),
                        resource_owner_key=os.getenv("ACCESS_TOKEN"), resource_owner_secret=os.getenv("ACCESS_TOKEN_SECRET"))

# OAuthAPI
url = 'https://api.twitter.com/1.1/search/tweets.json'
responce=twitter.get(
    url,
    params={"q":"","result_type":"result_type","count":100,"tweet_mode":"extended"})

# idjson
for status in responce.json()["statuses"]:
    #idjson
    print("\t".join([str(status["id"]),json.dumps(status)]))
```
